```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10, 10, 1000)
y = list(map(math.sin, x))

plt.figure(figsize=(15,5))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin (x)')
plt.show()
```

```python
def derivative_func (func, epsilon):
    '''
    Функция, которая вычисляет производную некоторой функции в точке x
    '''
    
    def df(x):
        return (func(x+epsilon) - func(x))/epsilon
    
    return df

df = derivative_func(math.sin, 0.0001)
```

```python
dy = [df(item) for item in x]
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, dy, x, y)
plt.xlabel('x')
plt.ylabel('sin и производная')
plt.show()
```
